```python
dataset_path = dataset.location
yaml_path = os.path.join(dataset_path, 'data.yaml')

with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Ép đường dẫn về tuyệt đối để YOLO không bị nhầm
data['train'] = os.path.join(dataset_path, 'train/images')
data['val'] = os.path.join(dataset_path, 'train/images')
data['test'] = os.path.join(dataset_path, 'train/images')

with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print(f"Đã cấu hình xong data.yaml tại: {yaml_path}")
```

```python
model = YOLO('yolov8s.pt')
# Thêm device=0 để dùng GPU T4
model.train(data=yaml_path, epochs=20, imgsz=640, plots=True, device=0)
```
